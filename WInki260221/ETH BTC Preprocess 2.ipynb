```python
import json, glob, pandas as pd, re, os

def process_crypto(currency):
    """Process and save tweets for one cryptocurrency"""
    print(f"\nProcessing {currency.upper()}...")
    
    tweets = []
    for file in glob.glob(f"CrypTop12/tweet/raw/{currency}/*.json"):
        with open(file) as f:
            for line in f:
                if line.strip():
                    d = json.loads(line)
                    tweets.append(d)
    
    df = pd.DataFrame(tweets)
    
    # Extract only needed columns
    df = pd.DataFrame({
        'date': pd.to_datetime(df['created_at']),
        'n_replies': df['replies_count'].astype(float),
        'n_likes': df['likes_count'].astype(float),
        'n_retweets': df['retweets_count'].astype(float),
        'text': df['tweet']
    })
    
    # Clean
    df = df[df['date'].dt.year.between(2017, 2022)]
    df = df[df['n_retweets'] > 0]
    df = df.drop_duplicates(subset=['date', 'text'])  # Simple duplicate removal
    df['text'] = df['text'].str.replace(r'http\S+', '', regex=True).str.strip()
    df = df[df['text'] != '']  # Remove empty tweets
    
    final_min = df['date'].min()
    final_max = df['date'].max()
    print(f"📅 Date range: {final_min} to {final_max}")
    
    # Save with semicolon separator, only 5 columns
    df.to_csv(f"{currency}_tweets_2017_2022.csv", index=False, sep=';')
    print(f"✅ Saved {len(df)} tweets to {currency}_tweets_2017_2022.csv")
    print(f"   Columns: {df.columns.tolist()}")
    return df

# Run for both
eth = process_crypto('eth')
btc = process_crypto('btc')

print(f"\n📊 Final: ETH={len(eth)} tweets, BTC={len(btc)} tweets")

# Show sample to verify format
print("\n📝 ETH Sample (first 3 rows):")
print(eth.head(3).to_string())

print("\n📝 BTC Sample (first 3 rows):")
print(btc.head(3).to_string())

# Verify the format by reading back the first few lines
print("\n🔍 Verifying file format:")
for currency in ['eth', 'btc']:
    filename = f"{currency}_tweets_2017_2022.csv"
    with open(filename, 'r') as f:
        first_line = f.readline().strip()
        second_line = f.readline().strip()
        print(f"\n{currency.upper()} file format:")
        print(f"  Header: {first_line}")
        print(f"  Sample: {second_line[:100]}...")
```
